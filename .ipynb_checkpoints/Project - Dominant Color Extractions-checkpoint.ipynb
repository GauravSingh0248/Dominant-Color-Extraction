```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from sklearn.cluster import KMeans
```

```python
img=cv2.imread("img.jpg") #it reads the the image in BGR format
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img=cv2.resize(img,(306,382))
print(img.shape)
```

```python
plt.imshow(img)
```

```python
X=img.reshape((-1,3)) # (m * 3) #  ...m(r*c) i hace fixed 3 and ,-1 tells to find automatically
print(X.shape)
```

```python
k=8
model=KMeans(n_clusters=k)
model.fit(X)
```

```python
centroids=model.cluster_centers_
```

```python
print(centroids)
```

```python
colors=np.array(centroids,dtype='uint8')
```

```python
print(colors)
```

```python
# #  Demo Color Patch (100,100,3) , to create a patch of -->Red Color

# mat=np.zeros((100,100,3),dtype='uint8')
# mat[:,:,2]=150 # making the channel one as whole red
# print(mat)
# plt.imshow(mat)
```

```python
i=1
for color in colors:
    # print(color)
    plt.subplot(1,k,i)
    plt.axis("off")
    i=i+1
    mat=np.zeros((100,100,3),dtype='uint8')
    mat[:,:,:]=color
    # print(mat)
    plt.imshow(mat)
plt.show()
```
